# Bagian 2: Data Preprocessing & Wrangling

Notebook ini mendemonstrasikan proses data cleaning (mengatasi missing values, data duplikat, outliers), feature scaling, dan categorical encoding menggunakan Pandas dan Scikit-Learn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

sns.set_theme(style="whitegrid")
```

## 1. Membuat Dummy Dataset Mentah

```python
data = {
    'Nama': ['Budi', 'Siti', 'Joko', 'Siti', 'Rini', 'Andi', 'Dewi', 'Joko'],
    'Umur': [25, 30, np.nan, 30, 45, 22, 28, np.nan],
    'Gaji': [5000000, 7000000, 6000000, 7000000, 25000000, 4500000, 5800000, 6000000],
    'Kota': ['Jakarta', 'Bandung', 'Jakarta', 'Bandung', 'Surabaya', 'Jakarta', 'Bandung', 'Jakarta']
}

df = pd.DataFrame(data)
print("Dataset Mentah:")
df
```

## 2. Mengatasi Duplikasi Data

```python
dup_count = df.duplicated().sum()
print(f"Jumlah data duplikat: {dup_count}")

df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print("DataFrame Setelah Duplikat Dihapus:")
df
```

## 3. Mengatasi Missing Values

```python
print("Jumlah missing values per kolom:")
print(df.isnull().sum())

# Mengimputasi (mengisi) nilai kosong pada Umur dengan nilai median
umur_median = df['Umur'].median()
df['Umur'].fillna(umur_median, inplace=True)

print(f"\nUmur kosong diimputasi dengan median ({umur_median}):")
df
```

## 4. Mendeteksi & Mengatasi Outliers (Pencilan)

Gaji Rini sebesar 25.000.000 terlihat sangat menyimpang dari karyawan lain. Mari kita gunakan metode IQR untuk mendeteksi outlier secara statistik.

```python
Q1 = df['Gaji'].quantile(0.25)
Q3 = df['Gaji'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (Kuartil 1): {Q1}")
print(f"Q3 (Kuartil 3): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Batas Atas Outlier: {upper_bound}")

# Deteksi baris yang berisi outlier
outliers = df[df['Gaji'] > upper_bound]
print("\nData Outlier Terdeteksi:")
outliers
```

```python
# Penanganan Outlier: Memfilter data di luar jangkauan outlier
df_clean = df[df['Gaji'] <= upper_bound].copy()
print("DataFrame Bersih dari Outliers:")
df_clean
```

## 5. Feature Scaling (Standardization & Normalization)

```python
# Inisialisasi Scaler
std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Standardisasi (Mean=0, Std=1)
df_clean['Gaji_Standardized'] = std_scaler.fit_transform(df_clean[['Gaji']])

# Normalisasi (Min=0, Max=1)
df_clean['Gaji_Normalized'] = minmax_scaler.fit_transform(df_clean[['Gaji']])

df_clean[['Nama', 'Gaji', 'Gaji_Standardized', 'Gaji_Normalized']]
```

## 6. Categorical Encoding

```python
# A. Label Encoding (Ordinal)
le = LabelEncoder()
df_clean['Kota_Encoded'] = le.fit_transform(df_clean['Kota'])

# B. One-Hot Encoding (Nominal)
df_final = pd.get_dummies(df_clean, columns=['Kota'], drop_first=True)

print("DataFrame Final Pasca Preprocessing:")
df_final
```
