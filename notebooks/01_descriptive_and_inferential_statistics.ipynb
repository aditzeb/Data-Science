# Bagian 1: Descriptive & Inferential Statistics

Notebook ini mendemonstrasikan analisis statistika dasar (deskriptif) menggunakan data dari proyek awal, kemudian memperluasnya ke dalam statistika inferensial (uji hipotesis) serta visualisasi distribusinya.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statistics

# Mengatur estetika plotting
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (9, 5)
```

## 1. Data Populasi & Statistika Deskriptif

Kita akan menggunakan data awal `populationData` yang merepresentasikan nilai ujian.

```python
populationData = [77, 50, 99, 60, 90, 82, 93, 75, 86, 99, 55, 79]
print("Dataset:", populationData)
```

### A. Ukuran Pemusatan Data (Mean, Median, Mode)

```python
x_mean = statistics.mean(populationData)
x_median = statistics.median(populationData)
x_mode = statistics.mode(populationData)

print("Rata-rata (Mean):", x_mean)
print("Nilai Tengah (Median):", x_median)
print("Modus (Mode):", x_mode)
```

### B. Ukuran Penyebaran Data (Standard Deviation & Variance)

Perlu diingat perbedaan antara populasi dan sampel:
- **Populasi:** Kita membagi dengan $N$ (seluruh anggota populasi). Parameter `ddof=0` di numpy.
- **Sampel:** Kita membagi dengan $N - 1$ (koreksi Bessel). Parameter `ddof=1` di numpy atau menggunakan modul `statistics`.

```python
# Deviasi Standar
std_pop = np.std(np.array(populationData))  # ddof=0 secara default di numpy
std_sample = statistics.stdev(populationData)  # ddof=1 secara default di statistics

print(f"Standard Deviation (Population): {std_pop:.4f}")
print(f"Standard Deviation (Sample): {std_sample:.4f}")
```

```python
# Varians
var_pop = np.var(np.array(populationData))  # ddof=0
var_sample = np.var(np.array(populationData), ddof=1)  # ddof=1

print(f"Variance (Population): {var_pop:.4f}")
print(f"Variance (Sample): {var_sample:.4f}")
```

```python
# Koefisien Variasi (CV) = (Std Dev Sampel / Mean) * 100
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100
print(f"Coefficient of Variation: {cv(populationData):.4f}%")
```

### C. Visualisasi Distribusi Data

```python
plt.figure(figsize=(8, 4.5))
sns.histplot(populationData, kde=True, color='royalblue', bins=6, edgecolor='black')
plt.axvline(x_mean, color='red', linestyle='--', linewidth=2, label=f'Mean ({x_mean:.2f})')
plt.axvline(x_median, color='forestgreen', linestyle='-', linewidth=2, label=f'Median ({x_median:.1f})')
plt.title('Distribusi Nilai Ujian (Descriptive)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

## 2. Statistika Inferensial (Uji Hipotesis)

Mari kita uji secara inferensial apakah rata-rata nilai ujian kelompok siswa ini berbeda secara signifikan dari standar kelulusan nasional yaitu **75**.

### Uji Hipotesis Satu Sampel (One-Sample t-Test)
- $H_0$: Rata-rata nilai siswa ($\mu$) = 75
- $H_1$: Rata-rata nilai siswa ($\mu$) $\neq$ 75
- Tingkat Signifikansi $\alpha$ = 0.05

```python
t_stat, p_val = stats.ttest_1samp(populationData, popmean=75)
print(f"t-Statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
```

```python
alpha = 0.05
if p_val <= alpha:
    print(f"P-value ({p_val:.4f}) <= alpha ({alpha}). TOLAK H0.")
    print("Kesimpulan: Rata-rata nilai ujian siswa berbeda signifikan dari 75.")
else:
    print(f"P-value ({p_val:.4f}) > alpha ({alpha}). GAGAL MENOLAK H0.")
    print("Kesimpulan: Secara statistik, tidak ada perbedaan signifikan antara rata-rata nilai siswa dengan 75.")
```
