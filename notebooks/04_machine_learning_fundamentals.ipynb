# Bagian 4: Machine Learning Fundamentals (Scikit-Learn)

Notebook ini melatih 3 jenis model machine learning dasar menggunakan Scikit-Learn:
1. **Supervised Regression:** Prediksi nilai kontinu
2. **Supervised Classification:** Klasifikasi kategori target
3. **Unsupervised Clustering:** Pengelompokan data tanpa label

Serta cara melakukan **Hyperparameter Tuning** dengan `GridSearchCV`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix, silhouette_score
from sklearn.datasets import make_regression, make_classification, make_blobs

sns.set_theme(style="whitegrid")
```

## 1. Supervised Learning: Regression (Regresi Linear)

Kita akan memprediksi suatu nilai kontinu berdasarkan 1 fitur input.

```python
# A. Generate Dataset Regresi
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# B. Split Train & Test
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# C. Latih Model
reg_model = LinearRegression()
reg_model.fit(X_train_r, y_train_r)

# D. Prediksi
y_pred_r = reg_model.predict(X_test_r)

# E. Evaluasi
mse = mean_squared_error(y_test_r, y_pred_r)
r2 = r2_score(y_test_r, y_pred_r)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R2 Score): {r2:.4f}")
```

```python
# Plot Garis Regresi
plt.figure(figsize=(7, 4))
plt.scatter(X_test_r, y_test_r, color='royalblue', label='Data Asli')
plt.plot(X_test_r, y_pred_r, color='tomato', linewidth=2.5, label='Garis Regresi')
plt.title('Evaluasi Model Regresi Linear', fontsize=12, fontweight='bold')
plt.xlabel('Fitur X')
plt.ylabel('Target Y')
plt.legend()
plt.show()
```

## 2. Supervised Learning: Classification (Random Forest)

Kita akan mengklasifikasikan data biner (0 atau 1) menggunakan `RandomForestClassifier`.

```python
# A. Generate Dataset Klasifikasi
X_clf, y_clf = make_classification(n_samples=500, n_features=5, n_informative=3, random_state=42)

# B. Split Train & Test
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# C. Latih Model
clf_model = RandomForestClassifier(n_estimators=50, random_state=42)
clf_model.fit(X_train_c, y_train_c)

# D. Prediksi
y_pred_c = clf_model.predict(X_test_c)

# E. Evaluasi
print("Akurasi Model:", accuracy_score(y_test_c, y_pred_c))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test_c, y_pred_c))
print("\nClassification Report:")
print(classification_report(y_test_c, y_pred_c))
```

## 3. Unsupervised Learning: Clustering (K-Means)

Mengelompokkan data yang tidak memiliki label ke dalam 3 cluster terpisah.

```python
# A. Generate Data 2-Dimensi
X_clust, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# B. Latih K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_clust)

# C. Evaluasi Silhouette Score
score = silhouette_score(X_clust, y_kmeans)
print(f"Silhouette Score (K=3): {score:.4f}")
```

```python
# Visualisasi Cluster
plt.figure(figsize=(7, 4.5))
plt.scatter(X_clust[:, 0], X_clust[:, 1], c=y_kmeans, s=30, cmap='viridis', alpha=0.8)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=150, label='Centroids')
plt.title('Hasil K-Means Clustering', fontsize=12, fontweight='bold')
plt.legend()
plt.show()
```

## 4. Hyperparameter Tuning dengan GridSearchCV

Mencari nilai hyperparameter `n_estimators` dan `max_depth` terbaik untuk model Random Forest Classifier kita.

```python
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [3, 5, 7]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_c, y_train_c)

print("Hyperparameter Terbaik:", grid_search.best_params_)
print(f"Akurasi Terbaik Hasil CV: {grid_search.best_score_:.4f}")
```
