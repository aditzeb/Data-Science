# Bagian 3: Exploratory Data Analysis (EDA) & Visualisasi

Notebook ini menunjukkan cara menjelajahi dataset secara komprehensif menggunakan grafik statistik interaktif dari library Matplotlib dan Seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set tema visual agar terlihat estetik dan premium
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 11
```

## 1. Membuat Sintetis Dataset Properti Perumahan

```python
np.random.seed(42)
n_samples = 300

luas = np.random.normal(120, 40, n_samples).clip(40, 300)
kamar = np.random.choice([1, 2, 3, 4, 5], size=n_samples, p=[0.1, 0.3, 0.4, 0.15, 0.05])
tipe = np.random.choice(['Subsidi', 'Komersil', 'Elite'], size=n_samples, p=[0.4, 0.4, 0.2])

# Harga dipengaruhi luas, jumlah kamar, dan tipe properti
harga = (luas * 3.5) + (kamar * 45) + np.random.normal(100, 50, n_samples)
harga[tipe == 'Subsidi'] *= 0.7
harga[tipe == 'Elite'] *= 1.8
harga = harga.clip(100, 2500) # rentang harga 100jt - 2.5m

df = pd.DataFrame({
    'Luas_Bangunan': luas,
    'Jumlah_Kamar': kamar,
    'Tipe_Properti': tipe,
    'Harga_Jual_Juta': harga
})
print("5 Data Teratas:")
df.head()
```

## 2. Univariate Analysis

Menganalisis sebaran nilai pada satu variabel tunggal.

```python
# A. Distribusi Harga Jual (Histogram + KDE)
fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))

sns.histplot(df['Harga_Jual_Juta'], kde=True, color='teal', ax=axes[0])
axes[0].set_title('Distribusi Harga Jual Properti', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Harga Jual (Juta Rupiah)')
axes[0].set_ylabel('Jumlah Properti')

# B. Distribusi Tipe Properti (Countplot)
sns.countplot(data=df, x='Tipe_Properti', palette='viridis', order=['Subsidi', 'Komersil', 'Elite'], ax=axes[1])
axes[1].set_title('Jumlah Properti Berdasarkan Tipe', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Tipe Properti')
axes[1].set_ylabel('Jumlah')

sns.despine()
plt.tight_layout()
plt.show()
```

## 3. Bivariate Analysis

Menganalisis hubungan timbal balik antara dua variabel.

```python
# A. Luas Bangunan vs Harga Jual (Scatter plot dengan tren)
fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))

sns.scatterplot(data=df, x='Luas_Bangunan', y='Harga_Jual_Juta', hue='Tipe_Properti', palette='Set2', alpha=0.8, ax=axes[0])
axes[0].set_title('Hubungan Luas Bangunan terhadap Harga Jual', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Luas Bangunan ($m^2$)')
axes[0].set_ylabel('Harga Jual (Juta Rupiah)')

# B. Boxplot Tipe Properti vs Harga Jual (Kategori vs Numerik)
sns.boxplot(data=df, x='Tipe_Properti', y='Harga_Jual_Juta', palette='Set2', order=['Subsidi', 'Komersil', 'Elite'], ax=axes[1])
axes[1].set_title('Harga Jual Berdasarkan Tipe Properti', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Tipe Properti')
axes[1].set_ylabel('Harga Jual (Juta Rupiah)')

sns.despine()
plt.tight_layout()
plt.show()
```

## 4. Multivariate Analysis

Menganalisis hubungan banyak variabel sekaligus.

```python
# Menghitung korelasi numerik
corr_matrix = df.corr(numeric_only=True)

# Heatmap korelasi
plt.figure(figsize=(6, 4.5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Matriks Korelasi Fitur Properti', fontsize=12, fontweight='bold', pad=15)
plt.show()
```

```python
# Pairplot korelasi visual
sns.pairplot(df, hue='Tipe_Properti', palette='Set1', diag_kind='kde')
plt.suptitle('Hubungan Berpasangan Fitur Properti (Pairplot)', y=1.02, fontsize=14, fontweight='bold')
plt.show()
```
